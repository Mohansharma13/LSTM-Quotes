```python
import numpy as np
import pandas as pd
```

```python
data_set=pd.read_csv("archive\Quotes Dataset.csv")
```

```python
print(data_set.shape)
data_set.head()
```

```python
quotes=data_set["Quote"]
raw_quotes=list(quotes)
raw_quotes
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer=Tokenizer()
```

```python
tokenizer.fit_on_texts(raw_quotes)
```

```python
tokenizer.word_index
```

```python
words=tokenizer.word_index
print("total number of word in our vocab",len(words))
```

```python
tokenized_quotes=[]
for sentance in raw_quotes:
    tokenized_sequence=tokenizer.texts_to_sequences([sentance])[0]
    for i in range(1,len(tokenized_sequence)):
        tokenized_quotes.append(tokenized_sequence[:i+1])
```

```python
tokenized_quotes
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_sequence=pad_sequences(tokenized_quotes,padding="pre")
```

```python

padded_sequence
```

```python
X=padded_sequence[:,:-1]
Y=padded_sequence[:,-1]
```

```python
print("shape of quotes ",X.shape)
X
```

```python
Y.shape
```

```python
# we are going to use muti class catigorical clasification of we have to one hot encode or targets
from tensorflow.keras.utils import to_categorical
targets = to_categorical(Y,num_classes=len(words)+1)#num classes is number of word , as word ranking in tokenizer start from 1 , so me have to add on 
```

```python
targets.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
model = Sequential()
model.add(Embedding(250, 100, input_length=23))
model.add(LSTM(150))
model.add(Dense(250, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
tf.config.list_physical_devices("GPU")
model.fit(X,targets,epochs=10)
```

```python
import time
text = "You only live once"

for i in range(7):
  # tokenize
  token_text = tokenizer.texts_to_sequences([text])[0]
  # padding
  padded_token_text = pad_sequences([token_text],maxlen=23, padding='pre')
  # predict
  pos = np.argmax(model.predict(padded_token_text))

  for word,index in tokenizer.word_index.items():
    if index == pos:
      text = text + " " + word
      print(text)
```
